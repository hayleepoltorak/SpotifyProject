```python
from collections import Counter 
print(Counter(df['Genre'].values))

#Counter package helps look at the frequency so we need to import that to sort our info
#prints the frequencies for each genre
```

```python
print(Counter(df['Genre'].values).most_common(5))
#this limits our data from only the top 5 genres that are most popular
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('Dance pop', 'pop', 'latin', 'canadian hip hop', 'Edm')  #the top 5 genres 
y_pos = np.arange(len(objects)) #makes the length of each bar in the graph
performance = [8,7,5,3,3] #entered the length of each corresponding genre

plt.barh(y_pos, performance, align='center', alpha=0.5) #handles the dimensions of the graph
plt.yticks(y_pos, objects)
plt.xlabel('frequency') #title of x axis on the graph
plt.title('Most Common Genres') #title of the graph

plt.show()
```

```python
def get_frequencies(column_name):
    print(Counter(df[column_name].values))
    print(dict(Counter(df[column_name].values).most_common(5)))
    bar_plot = dict(Counter(df[column_name].values).most_common(5))
    plt.bar(*zip(*bar_plot.items()))
    plt.show() 
#we created a function named get_frequencies in order to scrape the data from the source and get the data on the genres and their frequencies
```
